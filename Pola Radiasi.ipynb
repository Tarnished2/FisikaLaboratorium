```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel(r'D:\Dokumen\Kuliah\Kuliah\Semester 6\FisLab II\Optik\Pola Radiasi Antena\antengab.xlsx')
```

```python
# Dapatkan kolom sudut dan data
sudut = np.radians(data['sudut'])

data1 = data['antena 1']
data2 = data['antena 2']
data3 = data['antena 3']

# Buat polar plot
plt.polar(sudut, data1, label='Antena 1')
plt.polar(sudut, data2, label='Antena 2')
plt.polar(sudut, data3, label='Antena 3')

plt.title('Pola Radiasi Antena dalam Koordinat Polar')
plt.legend()
plt.show()
```

```python
# Dapatkan kolom sudut dan data
sudut = data['sudut']

data1 = data['antena 1']
data2 = data['antena 2']
data3 = data['antena 3']

# Buat polar plot
plt.plot(sudut, data1, label='Antena 1')
plt.plot(sudut, data2, label='Antena 2')
plt.plot(sudut, data3, label='Antena 3')

plt.title('Pola Radiasi Antena dalam Koordinat Kartesian')
plt.xlabel("Sudut (Derajat)")
plt.ylabel("Intensitas (dB)")
plt.grid(True)
plt.legend()
plt.show()
```
