```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = pd.read_excel(r'D:\Dokumen\Kuliah\Kuliah\Semester 6\FisLab II\Medis\TPS\data\rayhan2beam.xlsx')
```

```python
#2 Beam

#GTV
x_axis = file['_4']
y_axis = file['GTV Value (% of 85.382 cc)']
plt.plot(x_axis, y_axis, label='GTV')

#Lung Left
y_axis = file['Lung-Left Value (% of 2238.950 cc)']
plt.plot(x_axis, y_axis, label='Lung Left')

#Lung Right
y_axis = file['Lung-Right Value (% of 2477.200 cc)']
plt.plot(x_axis, y_axis, label='Lung Right')

#Esophagus
y_axis = file['Esophagus Value (% of 38.827 cc)']
plt.plot(x_axis, y_axis, label='Esophagus')

#Neoplasm
y_axis = file[' Primary Value (% of 85.382 cc)']
plt.plot(x_axis, y_axis, label='Neoplasm')

plt.xlabel("Dose (Gy)")
plt.ylabel("Volume (%)")
plt.grid(True)
plt.title('Dose Volume Histogram 2 Beam')
plt.legend()  # Menampilkan legenda
plt.show()
```
