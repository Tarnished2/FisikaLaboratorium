```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = pd.read_excel(r'C:\Users\GaMiN\Documents\Python\03 - Fisika Laboratorium\.vscode\data perhitungan konduktivitas.xlsx')
```

```python
# frekuensi terhadap konduktivitas rangkaian RC
x_axis = file['freq']
y_axis = file['Konduktivitas']
plt.plot(x_axis, y_axis)
plt.xlabel("Frekuensi")
plt.ylabel("Konduktivitas RC")
plt.grid(True)
plt.title('Grafik Frekuensi Terhadap Konduktivitas Rangkaian RC')
plt.show()
```
