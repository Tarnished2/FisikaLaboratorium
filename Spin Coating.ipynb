```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [1000, 1500, 2000]
y = [6000, 4000, 2000]
```

```python
slope1, intercept1, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
    return slope1 * x + intercept1

mymodel = list(map(myfunc, x))
plt.xlabel('Kecepatan Putar (rpm)')
plt.ylabel('Ketebalan Lapisan (px)')
plt.title('Grafik Hubungan Kecepatan Putaran dengan Ketebalan Lapisan Tipis')
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```
