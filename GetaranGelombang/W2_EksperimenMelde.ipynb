# Persamaan yang digunakan

## 1. Kecepatan Gelombang

$$ v = \frac{F}{\mu} $$

- Dimana $F$ adalah gaya tegangan tali dan $μ$ adalah rapat massa linear tali.

## 2. Gaya Tegangan Tali

$$ F=Mg $$
- Dimana $F$ adalah gaya tegangan tali, $M$ adalah massa total beban, dan $g$ adalah percepatan gravitasi.

## 3. Rapat Massa Linear Tali

$$ \mu = \frac{m}{L} $$
- Dimana $μ$ adalah rapat massa linear tali, $m$ adalah massa tali, dan $L$ adalah panjang tali.

## 4. Panjang Gelombang Ideal

$$ \lambda = \frac{1}{f} \sqrt{\frac{F}{μ}} $$
- Dimana $f$ adalah frekuensi vibrator, $F$ adalah gaya tegangan tali, $μ$ adalah rapat massa linear tali, dan $λ$ adalah panjang gelombang ideal gelombang stasioner.

## 5. Panjang Gelombang Bergantung Segmen

$$ λ=\frac{L}{n} $$
- Dengan $L$ adalah panjang gelombang percobaan dan $n$ adalah banyak gelombang.

## 1. Tali Kuning

```python
file_path = os.path.join(path, file_name)
data = pd.read_excel(file_path, sheet_name='kuning')
data
```

```python
m = 0.00142 #dalam kg
L = 1.5 #dalam meter
g = 9.8 #dalam m/s^2
```

### 1. Gaya Tegangan Tali

```python
data['Gaya Tegangan Tali'] = data['Beban'] * 9.8
data
```

### 2. Rapat Massa Linear Tali

```python
miu = m / L
print('miu = ', miu)
```

### 3. Kecepatan Gelombang

```python
data['Kecepatan Gelombang'] = data['Gaya Tegangan Tali'] / miu
data['Kecepatan Gelombang Kuadrat'] = (data['Gaya Tegangan Tali'] / miu)**2
data
```

### 4. Panjang Gelombang Ideal

```python
f = 50  # Hz
data['Panjang Gelombang Ideal'] = (1 / f) * (data['Gaya Tegangan Tali']/miu).apply(math.sqrt)
data
```

### 5. Panjang Gelombang Bergantung Segmen

```python
data['Panjang Gelombang Segmen'] = data['Panjang Gelombang Percobaan'] / data['Banyak Gelombang']
data
```

### 6. Plot Grafik

#### a. Hubungan Tegangan Tali dengan Kecepatan Gelombang Kuadrat Tali Kuning

```python
x = data['Kecepatan Gelombang Kuadrat']
y = data['Gaya Tegangan Tali']
plt.plot(x, y, color='yellow')
plt.xlabel('v kuadrat')
plt.ylabel('gaya tegangan tali')
plt.title('Grafik Hubungan Tegangan Tali dengan Kecepatan Gelombang Kuadrat Tali Kuning')
plt.show()
```

#### b. Hubungan Rapat Massa Linear dengan Kecepatan Gelombang Kuadrat Tali Kuning
