```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = pd.read_excel(r'C:\Users\GaMiN\Documents\Python\03 - Fisika Laboratorium\.vscode\akustik.xlsx')
```

```python
# frekuensi terhadap spl
x_axis = file['Freq,']
y_axis = file['g']
plt.plot(x_axis.astype(str), y_axis)
plt.xlabel("Frekuensi")
plt.ylabel("SPL")
plt.grid(True)
plt.title('Grafik Frekuensi Terhadap SPL')
plt.show()
```
